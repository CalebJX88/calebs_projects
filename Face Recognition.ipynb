# **Face Detection Using OpenCV**
A simple face detection model using OpenCV, where we will check for the n


### **Importing Libraries**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

### **Loading Image**


```python
img = cv2.imread('elon_musk.jpg')

# converting to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap = 'gray')
print('Image successfully loaded in grayscale')
```

### **Converting Image To RGB**


```python
def convert_to_RGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### **Haar Cascade Files**

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
```

### **Face Detection**

```python
faces_rects = haar_cascade_face.detectMultiScale(img_gray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

### **Drawing Rectangle Around Face**


```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(img, (x, y), (x+w, y+h), (0, 250, 0), 25)
#250 represents the width of the square
#25 represents the thickness of the square
```

```python
#printing the image that has been converted to RGB
plt.imshow(convert_to_RGB(img))
print('Image successfully converted back to RGB')
print('Square has been drawn to detect face')
```
